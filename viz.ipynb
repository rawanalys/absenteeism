```python
df_new.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(data=df_new, x="Age", y="Probability", errorbar=None) 
plt.suptitle("Age v/s Probability") 
plt.savefig('agePY1.jpeg')
```

```python
sns.jointplot(data=df_new, x="Age", y="Probability",height=7,ratio=2,kind="reg")
plt.suptitle("Age v/s Probability")
plt.savefig('agePY.jpeg')
```
